```python
# def createZoneTable(zone_factor,westlimit=-74.2635, southlimit=40.4856, eastlimit=-73.7526, northlimit=40.9596):

def createZoneTable(zone_factor,westlimit, southlimit, eastlimit, northlimit):
    zone_table = list()
    #Your code goes here
    zone_lat_diff = (northlimit - southlimit)/zone_factor
    zone_long_diff = (abs(westlimit - eastlimit))/zone_factor
#     print(zone_lat_diff,zone_long_diff)
    init_long = westlimit
    init_lat = southlimit
#     print(init_long + zone_long_diff)
#     print(init_lat + zone_lat_diff)
    counter = 0 
    total_list = []
    for i in range(0,zone_factor*zone_factor):

        row_long_new = init_long + zone_long_diff
        row_lat_new_constant = init_lat + zone_lat_diff
        
        first_coordinate = (init_long, init_lat)
        second_coordinate = (row_long_new, init_lat)
        
        third_coordinate = (row_long_new,row_lat_new_constant)
        fourth_coordinate = (init_long , row_lat_new_constant)
        fifth_coordinate = first_coordinate
        
        
        
        
        zone_list = [first_coordinate, second_coordinate, third_coordinate, fourth_coordinate, fifth_coordinate]

        zone_list_map = (zone_list)
        counter = counter + 1
        if(counter % zone_factor == 0):
            init_lat = init_lat + zone_lat_diff
            init_long = westlimit
        else:
            init_long = row_long_new
#         total_list.append(zone_list_map)
        total_list.extend(zone_list)
        lat_long_list.extend(zone_list)

    
    return total_list
```
